## Описание технологического процесса

После первичной обработки дроблёная смесь добытой руды проходит флотацию (обогащение) и двухэтапную очистку

1. **Флотация.**
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. **Очистка.**
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Технологический процесс
- `Rougher feed` — исходное сырье
- `Rougher additions` (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
   - `Xanthate` — ксантогенат (промотер, или активатор флотации);
   - `Sulphate` — сульфат (на данном производстве сульфид натрия);
   - `Depressant` — депрессант (силикат натрия).
- `Rougher process` (англ. «грубый процесс») — флотация
- `Rougher tails` — отвальные хвосты
- `Float banks` — флотационная установка
- `Cleaner process` — очистка
- `Rougher Au` — черновой концентрат золота
- `Final Au` — финальный концентрат золота

### Параметры этапов
- `air amount` — объём воздуха
- `fluid levels` — уровень жидкости
- `feed size` — размер гранул сырья
- `feed rate` — скорость подачи

### Наименование признаков
Наименование признаков имеет вид:

**`[этап].[тип_параметра].[название_параметра]`**

Возможные значения для блока **`[этап]`**:
- `rougher` — флотация
- `primary_cleaner` — первичная очистка
- `secondary_cleaner` — вторичная очистка
- `final` — финальные характеристики

Возможные значения для блока **`[тип_параметра]`**:
- `input` — параметры сырья
- `output` — параметры продукта
- `state` — параметры, характеризующие текущее состояние этапа
- `calculation` — расчётные характеристики

### Расчёт эффективности

Эффективность обогащения рассчитывается по формуле <a id='eq_recovery'></a>
$$Recovery = \frac{C \cdot (F-T)}{F \cdot (C-T)} \cdot 100 \%, $$
где 
- $C$ — доля золота в концентрате после флотации/очистки;
- $F$ — доля золота в сырье/концентрате до флотации/очистки;
- $T$ — доля золота в отвальных хвостах после флотации/очистки.

### Метрика качества
Для решения задачи используется метрика качества — *sMAPE* (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).

*sMAPE* вычисляется по формуле: <a id='eq_smape'></a> 
$$ sMAPE = \sum_{i=1}^N \frac {2 |y_i-\hat{y_i}|} {|y_i|+|\hat{y_i}|} \cdot 100 \%, $$
где 
- $y_i$ — значение целевого признака для объекта с порядковым номером $i$ в выборке, на которой измеряется качество.;
- $\hat{y_i}$ — значение предсказания для объекта с порядковым номером $i$, например, в тестовой выборке;
- $N$ — количество объектов в выборке.

### Целевые переменные

Нужно спрогнозировать сразу две величины:
1. эффективность обогащения чернового концентрата **`rougher.output.recovery`**;
2. эффективность обогащения финального концентрата **`final.output.recovery`**.

Итоговая метрика складывается из двух величин:  <a id='eq_smape_final'></a> 
$$ sMAPE_{итоговое} = 25 \% \cdot sMAPE(roughter) + 75 \% \cdot sMAPE(final)$$

```python
# Диаграммы размаха и гистограммы для одного параметра из двух датасетов
def show_hists(df1, df2, col, bins=100, xlim=None, ylim=None, figsize=None,
    title = 'Распределение значений', 
    xlabel = None, ylabel = 'Количество значений'):
    '''
    Функция построения гистограмм и диаграмм размаха для столбца col датафреймов df1 и df2
    
    Параметры 
    ---
    `bins` : int - задаёт число корзин для гистограммы, по умолчанию 100
    `title` : string - задаёт название графика
    `xlabel, ylabel` : string - подписи к осям
    `xlim, ylim, figsize` : параметры matplotlib 
    '''
    # создаём 3 сабплота, указываем для них соотношение занимаемой высоты
    _, (ax_box1, ax_box2, ax_hist) = plt.subplots(3, sharex=True, 
        gridspec_kw={"height_ratios": (.1, .1, .8)}, figsize=figsize)
    
    if col in df1.columns and col in df2.columns:
        # добавляем диаграммы размаха, рисуем каждый на отдельном полотне ax_box{i}
        sns.boxplot(x = df1[col], ax=ax_box1, color='steelblue')
        sns.boxplot(x = df2[col], ax=ax_box2, color='darkorange')
        # добавляем гистограммы с функциями распределения на одном полотне ax_hist
        sns.histplot(df1[col], ax=ax_hist, bins=bins, kde=True, color='steelblue')
        sns.histplot(df2[col], ax=ax_hist, bins=bins, kde=True, color='darkorange')
        
        if xlabel is None:  xlabel = col
        # удалим подпись графика для boxplot, чтобы не дублировалась
        ax_box1.set(xlabel='')
        ax_box2.set(xlabel='')
        # передаём параметры
        ax_hist.set(xlabel=xlabel, ylabel=ylabel, xlim=xlim, ylim=ylim)
        ax_hist.grid(visible=True)
        plt.suptitle(title, y = 0.92, fontsize=14)
        # добавляем подписи для легенды
        try: 
            plt.legend([df1.name, df2.name], fontsize='large')
        except: 
            plt.legend(['df1', 'df2'], fontsize='large')
    else:
        print('Столбец', col, 'отсутсвует в каком-то из датафреймов')


# вспомогательная функция для сортировки словаря по ключу или значению
def sort_dict(dictionary, ind=1):
    ''' Возвращает dict отсортированный словарь

        `dictionary` : dict - словарь для сортировки
        `ind` : int (0 или 1). 0: сортировка по ключу, 1: по значению
    '''
    return dict(sorted(dictionary.items(), key=lambda item: item[ind]))
```

### Проверка значений эффективности обогащения

Проверим расчётные значения эффективности обогащения для этапа флотации по [формуле для расчёта Recovery](#eq_recovery), в которой в нашем случае:
- $C$ = `rougher.output.concentrate_au` — доля золота в концентрате после флотации;
- $F$ = `rougher.input.feed_au` — доля золота в сырье до флотации;
- $T$ = `rougher.output.tail_au` — доля золота в отвальных хвостах после флотации.

```python
rougher_output_recovery_train = (100 * 
    ( df_train['rougher.output.concentrate_au'] * 
      (df_train['rougher.input.feed_au'] - df_train['rougher.output.tail_au'])
    ) /
    ( df_train['rougher.input.feed_au'] * 
      (df_train['rougher.output.concentrate_au'] - df_train['rougher.output.tail_au']) 
    ) )
print('MAE между расчётами и значениями датасета:',
      mean_absolute_error(rougher_output_recovery_train, df_train['rougher.output.recovery']))
del rougher_output_recovery_train
```

### Предобработка данных

Ниже предствалена схема процесса очистки с указанием соотвествия признаков в данных каждому этапу. Синим обозначены признаки, доступные в тестовой выюорке. На них будут обучаться модели, при этом для предсказания `rougher.output.recovery` будут использоваться только признаки `rougher.input` и `rougher.state`, для прелсказания `final.output.recovery` - все признаки с типом `input` и `state`

![technology.png](technology.png)

Отсортируем для удобства признаки в порядке этапов технологического процесса, также соберём названия признаков, которые будут использования для обучения моделей

```python
# словарь суффиксов признаков [этап].[тип_признака]
stages = {
    '0' : 'rougher.input',
    '1' : 'rougher.state',
    '2' : 'rougher.output',
    '3' : 'rougher.calculation',
    '4' : 'primary_cleaner.input',
    '5' : 'primary_cleaner.state',
    '6' : 'primary_cleaner.output',
    '7' : 'secondary_cleaner.state',
    '8' : 'secondary_cleaner.output',
    '9' : 'final.output',
}

# сфорируем словарь с короткими вариантами наименования признаков
cols_short = {}
for col in df_train.columns:
    for s in stages:
        if stages[s] in col:
            cols_short[col] = col.replace(stages[s], s)
cols_short = sort_dict(cols_short)

# отсортируем
df_train = df_train.reindex(cols_short.keys(), axis=1)
df_test = df_test.reindex(
    sort_dict({col : cols_short[col] for col in df_test.columns}).keys(), axis=1)

# список признаков для предсказания восстановления золота после флотации 
features_rougher = []
for f in cols_short:
    if 'rougher.input' in f or 'rougher.state' in f:
        features_rougher.append(f)
features_rougher.append('rougher.output.recovery')

# список признаков для предсказания воссатновления золота после всех этапов 
features_final = []
for f in cols_short:
    if 'input' in f or 'state' in f:
        features_final.append(f)
features_final.append('final.output.recovery')
```

Выведем сравнительную информацию на тренировочном и тестовом датасетах о признаках для предсказания восстановления золота после флотации, со средними распределениями `rougher.output.recovery`

## Модель

### Функция для вычисления `sMAPE`

Реализуем функции для расчёта [симметричного среднего абсолютного отклонения `sMAPE`](#eq_smape) и [итогового `sMAPE`](#eq_smape_final)

```python
def smape(y_true, y_pred):
    ''' Функция вычисления smape на известных и предсказанных значениях
    '''
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    summaries = 2*(np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred)))
    return np.mean(summaries)

def smape_result(smape_rougher, smape_final):
    return 0.25*smape_rougher + 0.75*smape_final

smape_score = make_scorer(smape, greater_is_better=False)
```

Выделим признаки и метки

```python
print('Train : missing', df_train.isna().sum().sum())
print('Test  : missing', df_test.isna().sum().sum())

X_train_rougher = df_train[features_rougher[:-1]].values
y_train_rougher = df_train[features_rougher[-1]].values
X_test_rougher  = df_test[features_rougher[:-1]].values
y_test_rougher  = df_test[features_rougher[-1]].values

X_train_final = df_train[features_final[:-1]].values
y_train_final = df_train[features_final[-1]].values
X_test_final  = df_test[features_final[:-1]].values
y_test_final  = df_test[features_final[-1]].values

print('Train shapes | rougher :', X_train_rougher.shape, y_train_rougher.shape,
      '| final :',  X_train_final.shape, y_train_final.shape)
print('Test shapes | rougher :', X_test_rougher.shape, y_test_rougher.shape,
      '| final :',  X_test_final.shape, y_test_final.shape)
```

### Линейная регрессия

Вычислим `sMAPE` на предсказаниях линейной регрессии, применив кросс-валидацию с разбиением датасета на 5 частей.

```python
smapes_r = -cross_val_score(LinearRegression(), X_train_rougher, y_train_rougher, cv=5, scoring=smape_score)
smapes_f = -cross_val_score(LinearRegression(), X_train_final, y_train_final, cv=5, scoring=smape_score)

print('Rougher sMAPES : mean = {:.3%}. \ncv: {}'.format(
      smapes_r.mean(), smapes_r))
print('Final sMAPES   : mean = {:.3%}. \ncv: {}'.format(
      smapes_f.mean(), smapes_f))

smapes_res = smape_result(smapes_r, smapes_f)
print('\nResult sMAPES  : mean = {:.3%}. \ncv: {}'.format(
      smapes_res.mean(), smapes_res))
```

### Масштабирование признаков

Посмотрим, как меняется метрика, если предварительно отмасштабировать признаки

Масштабирование не повлияло на качество метрики

### Решающие деревья

Произведём поиск наилучших гиперпараметров с помощью `GridSearchCV` на сетке параметров `min_samples_split` и `max_depth` регрессора `DecisionTreeRegressor`

```python
%%time
params = {'min_samples_split': range(2,11,2), 'max_depth': range(1,11)}

gdt_r = GridSearchCV(DecisionTreeRegressor(random_state=RANDOM_STATE), 
                     param_grid=params, scoring=smape_score, cv=5, n_jobs=8)
gdt_r.fit(X_train_rougher, y_train_rougher)

gdt_f = GridSearchCV(DecisionTreeRegressor(random_state=RANDOM_STATE), 
                     param_grid=params, scoring=smape_score, cv=5, n_jobs=8)
gdt_f.fit(X_train_final, y_train_final)

print('Rougher sMAPE: {:.3%}. Params {}'.format(-gdt_r.best_score_, gdt_r.best_params_))
print('Final sMAPE  : {:.3%}. Params {}'.format(-gdt_f.best_score_, gdt_f.best_params_))
print('\nResult sMAPE : {:.3%}\n'.format(-smape_result(gdt_r.best_score_ , gdt_f.best_score_)))
```

### Случайный лес

Аналогично с помощью кроссвалидации найдём наиулучшую модель случайного леса `RandomForestRegressor` на сетке параметров `max_depth`, `n_estimators`

```python
%%time
params = {'n_estimators': range(10,101,10), 'max_depth': range(2,11,2)} 

grf_r = GridSearchCV(RandomForestRegressor(random_state=RANDOM_STATE), 
                     param_grid=params, scoring=smape_score, cv=5, n_jobs=8)
grf_r.fit(X_train_rougher, y_train_rougher)

grf_f = GridSearchCV(RandomForestRegressor(random_state=RANDOM_STATE), 
                     param_grid=params, scoring=smape_score, cv=5, n_jobs=8)
grf_f.fit(X_train_final, y_train_final)

print('Rougher sMAPE: {:.3%}. Params {}'.format(-grf_r.best_score_, grf_r.best_params_))
print('Final sMAPE  : {:.3%}. Params {}'.format(-grf_f.best_score_, grf_f.best_params_))
print('\nResult sMAPE : {:.3%}\n'.format(-smape_result(grf_r.best_score_ , grf_f.best_score_)))
```

### Проверка лучшей модели

Вычислим sMAPE на тестовой выборке для случайного леса.

```python
model_r = RandomForestRegressor(random_state=RANDOM_STATE, **grf_r.best_params_, n_jobs=8) 
model_f = RandomForestRegressor(random_state=RANDOM_STATE, **grf_f.best_params_, n_jobs=8) 

model_r.fit(X_train_rougher, y_train_rougher)
model_f.fit(X_train_final, y_train_final)

y_pred_rougher = model_r.predict(X_test_rougher)
y_pred_final = model_f.predict(X_test_final)

smape_r = smape(y_test_rougher, y_pred_rougher)
smape_f = smape(y_test_final, y_pred_final)
print('Result sMAPE : {:.3%}'.format(smape_result(smape_r, smape_f)))
```

Проверим адекватность модели, сравнив со значением `sMAPE` на ответах модели, предсказывающей среднее значение.

```python
dr_r = DummyRegressor(strategy='mean').fit(X_train_rougher, y_train_rougher)
dr_f = DummyRegressor(strategy='mean').fit(X_train_final, y_train_final)

y_pred_rougher = dr_r.predict(X_test_rougher)
y_pred_final = dr_f.predict(X_test_final)

smape_r = smape(y_test_rougher, y_pred_rougher)
smape_f = smape(y_test_final, y_pred_final)
print('Result sMAPE : {:.3%}'.format(smape_result(smape_r, smape_f)))
```
