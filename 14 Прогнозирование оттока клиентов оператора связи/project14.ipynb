# Телеком

Оператор связи __«Ниединогоразрыва.ком»__ хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

**Описание услуг**

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. *digital subscriber line*) или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

```python
class hm():
    def dataframe_info(df):
        ''' Вывод информации о датафрейме `df`. Функционал схож с методом `pd.DataFrame.info()`

        Дополнительно: вывод процента заполненных (non-null) значений, количестве дублей
        '''
        info = (pd.DataFrame({
            'notNA': df.count(), 'notNA, %': df.count()/df.shape[0]*100,
            'dtype': df.dtypes})
            .style.bar(subset='notNA, %', vmin=0, color='lightblue')
            .format("{:.2f}", subset=['notNA, %'])
        )
        print('DataFrame shape  : {} rows, {} columns'.format(
            df.shape[0], df.shape[1]))
        print('Memory usage     : {:.2f} MB'.format(
            df.memory_usage().sum()/1024/1024))
        print('Duplicates count :', df.duplicated().sum())
        return info
    
    
    def plot_value_counts(series, n_values=25, fillna='NONE', figwidth=15, bar_thickness=0.5, 
                          verbose=True, show_percents=False):
        ''' Визуализация количества встречающихся значений в pd.Series

        Параметры
        ---
            `series` : pd.Series
            `column` : str - название столбца
            `n_values` : int - максимальное количество значений для отображения на диаграмме
            `fillna` : Any - значение, которым необходимо заполнить пропуски
            `figwidth` : float - ширина фигуры графика
            `bar_thickness` : float - ширина бара
            `verbose`: bool - показывать ли уникальные значения
            `show_percents`: bool - показывать долю значений в процентах
        '''
        _ = series.dropna().unique()
        if verbose:
            print('`{}`, {} unique values: \n{}'.format(series.name, len(_), sorted(_)))

        val_counts = series.fillna(fillna).value_counts()
        bar_values = val_counts.values[:n_values]
        bar_labels = val_counts.index[:n_values].astype('str')
        plt.figure(figsize=(figwidth, bar_thickness * min(len(val_counts), n_values)))
        ax = sns.barplot(x=bar_values, y=bar_labels)
        ax.set(title='"{}" value counts ({} / {})'
            .format(series.name, len(bar_labels), val_counts.shape[0]),
            #xlim=[0, 1.07*bar_values.max()]
            )
        if show_percents:
            labels = [f'{w/val_counts.values.sum()*100:0.1f}%' 
                        if (w := v.get_width()) > 0 else '' for v in ax.containers[0]]
        else:
            labels = bar_values
        plt.bar_label(ax.containers[0], labels=labels, label_type='center')
        for i in range(len(bar_labels)):
            if bar_labels[i] == fillna:
                ax.patches[i].set_color('black')
        plt.show()


    def plot_hist_and_box(series_list: list, bins=100, figsize=(12, 8), legend=None,
                        xlim=None, ylim=None,
                        xlabel=None, ylabel='Values counts'):
        ''' Функция построения гистограмм и диаграмм размаха для pd.Series из series_list
        
        Параметры 
        ---
            `bins` : int - задаёт число корзин для гистограммы, по умолчанию 100
            `xlabel, ylabel` : string - подписи к осям, если необходимо указать свои
            `xlim, ylim, figsize` : параметры matplotlib 
        '''
        if not (type(series_list) == list or type(series_list) == tuple):
            print("series should be list/tuple of pd.Series")
            return

        height_ratios = [.1]*len(series_list)
        height_ratios.append(1-sum(height_ratios))
        _, axs = plt.subplots(len(series_list)+1, sharex=True, figsize=figsize,
                            gridspec_kw={"height_ratios": height_ratios})
        cmap = plt.get_cmap('tab10')#'brg')
        names = ', '.join(set([s.name for s in series_list]))

        axs[0].set(title="Distribution of %s" % names)
        for i in range(len(series_list)):
            # диаграмма размаха
            sns.boxplot(x=series_list[i], ax=axs[i], color=cmap(i/10))
            axs[i].set(xlabel=series_list[i].name)
            # гистограмма с функцией распределения
            sns.histplot(series_list[i], ax=axs[-1], bins=bins, kde=True,
                        color=cmap(i/10))
        if not xlabel:
            xlabel = names
        axs[-1].set(xlabel=xlabel, ylabel=ylabel, xlim=xlim, ylim=ylim)
        axs[-1].grid(visible=True)

        if legend:
            plt.legend(legend, fontsize='large')
        plt.show()


    def best_cv_models(grid, count, score='score'):
        ''' Выводит таблицу с показателями моделей, показавших наилучшие значения метрики на кроссвалидации.

        Принимает  
            `grid` : результат GridSearchCV после fit(), 
            `count` : количество лучших моделей для вывода
            `score` : метрика
        Возвращает : pd.DataFrame c параметрами моделей
        '''
        print('Estimator: {}'.format(grid.estimator))
        print('Tested {} models. Splits: {}'.format(
            len(grid.cv_results_['params']), grid.cv))
        print('Best score = {}\n'.format(grid.best_score_))

        results = {}
        for s in set(['score', score]):
            if 'rank_test_'+s in grid.cv_results_.keys():
                best_idx = grid.cv_results_['rank_test_'+s].argsort()[:count]
            if 'mean_test_'+s in grid.cv_results_.keys():
                results['test '+score] = grid.cv_results_['mean_test_'+s][best_idx]
            if 'mean_train_'+s in grid.cv_results_.keys():
                results['train '+score] = grid.cv_results_['mean_train_'+s][best_idx]
        results['fit time, s'] = grid.cv_results_['mean_fit_time'][best_idx]
        results['score time, s'] = grid.cv_results_['mean_score_time'][best_idx]

        return pd.DataFrame(results).join(
            pd.DataFrame([grid.cv_results_['params'][i] for i in best_idx]))


    def roc_curve_plots(models, X_test, y_test, title='ROC Curve', figsize=(12, 9)):
        ''' Функция построения ROC кривой для моделей из переданнойго словаря

            Принимает:
        '''
        plt.figure(figsize=figsize)
        for name, model in models.items():
            pred = model.predict_proba(X_test)[:, 1]
            fpr, tpr, thresholds = roc_curve(y_test, pred)
            roc_auc = auc(fpr, tpr)
            plt.title(title)
            plt.plot(fpr, tpr, label='AUC = %0.4f (%s)' % (roc_auc, name))
        plt.legend(loc='lower right')
        plt.plot([0, 1], [0, 1], 'r--')
        plt.grid()
        plt.xlim([0, 1]), plt.ylim([0, 1])
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
        plt.show()

    def test_model(model, X_train, X_test, y_train, y_test, score_func=None):
        ''' 
        - Обучение модели `model` на выборках `X_train`, `y_train`
        - Предсказание обученной модели на наборе `X_test`
        - Вычисление метрики `score_func` на полученных предсказаниях и наборе `y_test`
        
        `score_func` : sklearn.metrics  

        Возвращает score, время обучения, время пердсказаний и сами предсказания в виде словаря
        '''
        # обучение
        t_beg = time.time()
        model.fit(X_train, y_train)
        time_fit = time.time() - t_beg
        # предсказания
        t_beg = time.time()
        y_pred = model.predict(X_test)
        time_predict = time.time() - t_beg
        # метрика
        score = score_func(y_test, y_pred)

        return {'fit time': time_fit,
                'predict time': time_predict,
                'score': score,
                'predictions': y_pred}

# Константы
RANDOM_STATE = 123
UPLOAD_DATE = '2020-02-01'
```

### Предобработка данных

Объеденим все таблицы в один датасет

```python
df = pd.concat(data, axis=1).droplevel(0, axis=1)
hm.dataframe_info(df)
```

Как и следовало ожидать, после объединения возникли пропуски

Объеденим все таблицы в одну и произведём предобработку данных.  
Проделаем следующие действия с полями:
1. Добавим целевую метку `Churn` - факт ухода клиента на основе поля `EndDate`
2. Поля `BeginDate` и `EndDate`:
    - изменим тип на datetime, пропуски в `EndDate` заменим на дату выгрузки данных
    - вычислим на основе этих полей значения периода пользования клиента услугами `UsageTime` ~~в месяцах~~ в днях!
3. Избавимся от пропусков в `TotalCharges`, преобразуем тип данных
4. Обработаем категориальные признаки
    - Возникшие пропуски в бинарных категориях заменим на новую, третью категорию, назовём её _'N/A'_
    - Категориальные признаки с более чем двумя категориями оставим без изменений
    - Явно укажем тип таких полей как категориальный 

```python
cat_features = (
    ['Type', 'PaymentMethod', 'PaperlessBilling'] + 
    list(data['personal'].columns) + 
    list(data['internet'].columns) + 
    list(data['phone'].columns)
)

# Вычисляем столбец с целевой меткой
df['Churn'] = (df['EndDate'] != 'No').astype(int)

# преобразуем тип данных полей с датой
df['BeginDate'] = pd.to_datetime(df['BeginDate'])
df.loc[df.EndDate == 'No', 'EndDate'] = UPLOAD_DATE 
df['EndDate'] = pd.to_datetime(df['EndDate'])

# вычисляем длительность действия договора в месяцах
df['UsageTime'] = (df['EndDate'] - df['BeginDate']).dt.days

# преобразуем поле TotalCharges в числовое, избавляясь от пропусков
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(0., inplace=True)

# заполним пропуски
for f in cat_features:
    df[f] = df[f].fillna('N/A')

hm.dataframe_info(df)
```

Выведем среднюю продолжительность использования услуг для обеих групп с разбивкой по признакам

```python
def grouped_means(f, cat_f):
    ''' Средние значения признака f для каждой категории из столбца cat_f
        для обеих групп из целевого признака Churn в датасете df
    '''
    val_counts = df[f].value_counts()
    plt.figure(figsize=(15, 0.5*len(val_counts)))
    if 'Yes' in val_counts.index:
        order = ['Yes','No']
        if 'N/A' in val_counts.index:
            order.append('N/A')
    else:
        order = None
    sns.barplot(x=df[cat_f], y=df[f].astype('str'),
                color='lightgrey', errcolor='grey', order=order)
    sns.barplot(x=df[cat_f], y=df[f].astype('str'),
                hue=df['Churn'], order=order)
    plt.xlabel("")
    plt.show()

display(Markdown('#### Средняя продолжительность использования услуг провайдера'))
for f in cat_features:
    grouped_means(f, 'UsageTime')
```

- У пользователей телефонии это время самое короткое
- Дополнительные услуги подключены преимущественно у постоянных клиентов
- Срок пользования услугами у абонентов с ежемесячной оплатой самый короткий

В такой же разбивке посмотрим на процентные соотношения

```python
def cat_features_counts(f, order=None):
    ''' Количество значений объектов категориального признака f для обеих групп 
        из целевого признака Churn в датасете df с отображением доли значений
    '''
    val_counts = df[f].value_counts()
    plt.figure(figsize=(15, 0.5*len(val_counts)))
    if 'Yes' in val_counts.index:
        order = ['Yes','No']
        if 'N/A' in val_counts.index:
            order.append('N/A')
    else:
        order = None
    ax = sns.countplot(y=df[f].astype('str'), hue=df['Churn'], dodge=False, order=order)
    labels0 = [f'{(v0.get_width()-v1.get_width())/v0.get_width():0.2f}' for (v0, v1)
               in zip(ax.containers[0], ax.containers[1])]
    labels1 = [f'{v1.get_width()/v0.get_width():0.2f}' for (v0, v1)
               in zip(ax.containers[0], ax.containers[1])]
    plt.bar_label(ax.containers[0], labels=labels0, padding=-36)
    plt.bar_label(ax.containers[1], labels=labels1, label_type='center')
    plt.show()

display(Markdown('#### Соотношения количества клиентов в зависимости от признаков'))
for f in cat_features:
    cat_features_counts(f)
```

## Обучение моделей

- Разделим на обучающую и тренировочную выборки в соотношении 4:1 
- Отделим целевой признак
- Отмасштабируем численные призхнаки

```python
# разделяем датасет
x_train, x_test, y_train, y_test = train_test_split(
    df.drop('Churn', axis=1), df['Churn'], test_size=0.2, 
    random_state=RANDOM_STATE #, stratify=y, shuffle=True # со стратификацией результаты хуже
) 
print('Обучающая выборка: ', x_train.shape, y_train.shape)
print('Тренировочная выборка: ', x_test.shape, y_test.shape)

# подгоняем скейлер на тренировочной выборке и масштабируем её численные признаки
scaler = StandardScaler()
x_train[numeric_features] = pd.DataFrame(
    scaler.fit_transform(x_train[numeric_features]),
    columns=numeric_features, index=x_train.index
)
# масштабируем численные признаки в тестовой выборке
x_test[numeric_features] = pd.DataFrame(
    scaler.transform(x_test[numeric_features]),
    columns=numeric_features, index=x_test.index
)
```

По требованию заказчика для оценки качества модели будем использовать метрику ROC AUC.  
Для более простой интерпретации результатов для отобранных моделей также посчитаем accuracy
Для обучения логистической регрессии и дерева решений закодируем категориальные признаки методом OHE, в LGBMClassifier и CatBoostClassifier будем передавать напрямую категориальные признаки  

Обучать будем с использованием кроссвалидации с разбиением датасета на 4 части. Запишем общие параметры в словать

```python
cv_params = {'scoring': 'roc_auc',
             'return_train_score': True,
             'cv': 4}
```

### Логистическая регрессия LogisticRegression

```python
%%time
pipe = Pipeline([
    ('ohe', ColumnTransformer(
        transformers=[('ohe', OneHotEncoder(sparse=False, drop='first'), cat_features)],
        verbose_feature_names_out=False, remainder='passthrough')),
    ('clf', LogisticRegression(random_state=RANDOM_STATE, max_iter=500))
])
params = {'clf__solver': ['lbfgs', 'liblinear'],
          'clf__C': np.logspace(-3, 3, 7),
          'clf__class_weight' : [None, 'balanced']
         } 
grid_lr = GridSearchCV(pipe, params, **cv_params, n_jobs=-1)
grid_lr.fit(x_train, y_train)

hm.best_cv_models(grid_lr, 10, score='AUC')
```

### Случайный лес RandomForestClassifier

Вместо перебора гипермараметров по всей сетке будем использовать рандомизированный перебор для ускорения обучения

```python
%%time
pipe = Pipeline([
    ('ohe', ColumnTransformer(
        transformers=[('ohe', OneHotEncoder(sparse=False, drop='first'), cat_features)],
        verbose_feature_names_out=False, remainder='passthrough')),
    ('clf', RandomForestClassifier(random_state=RANDOM_STATE))
])
params = {'clf__max_depth': range(4, 21, 4),
          'clf__min_samples_leaf': range(1, 11, 1),
          'clf__n_estimators': [10, 50, 100],
          'clf__class_weight' : [None, 'balanced']
         } 
grid_rf = RandomizedSearchCV(pipe, params, **cv_params, random_state=RANDOM_STATE, n_jobs=-1)
grid_rf.fit(x_train, y_train)

hm.best_cv_models(grid_rf, 10, score='AUC')
```

### Лес с бустингом LGBMClassifier

```python
%%time
params = {'learning_rate': [0.0005, 0.001, 0.01, 0.05, 0.1],
          'n_estimators': [100, 200, 300, 500],
          'num_leaves': [31, 71, 101, 151],
          'class_weight': [None, 'balanced']
          }
grid_lgbm = RandomizedSearchCV(
    LGBMClassifier(metric='auc', random_state=RANDOM_STATE), 
    params, **cv_params, random_state=RANDOM_STATE)

grid_lgbm.fit(x_train, y_train)
grid_lgbm.predict(x_test)

display(hm.best_cv_models(grid_lgbm, 10, score='AUC'))
```

### Лес с бустингом CatBoostClassifier

```python
%%time
params = {'max_depth': [4, 8],
          'learning_rate': [0.005, 0.01, 0.1],
          'n_estimators': [100, 200]
          }
grid_ctb = RandomizedSearchCV(
    CatBoostClassifier(random_state=RANDOM_STATE,
                       cat_features=cat_features,
                       verbose=False, task_type='GPU'), 
    params, **cv_params, n_iter=6, random_state=RANDOM_STATE)
grid_ctb.fit(x_train, y_train)
hm.best_cv_models(grid_ctb, 10, score='AUC')
```

### Результаты кроссвалидации

На кроссвалидации лучшуб метрику AUC-ROC выдал CatBoost, LGBM показывает признаки переобучения, но метрика также высокая и обучается намного быстрее.  
Построим диаграмму важности признаков

```python
model = grid_lgbm.best_estimator_.fit(x_train, y_train)
feature_importance = pd.Series(model.feature_importances_,
                               index=x_train.columns)
feature_importance = (feature_importance/feature_importance.max()).sort_values(ascending=False)
plt.figure(figsize=(15,8))
sns.barplot(x=feature_importance.values, y=feature_importance.index)
plt.title('Feature Importance', fontsize=16)
plt.xlabel('Importance', fontsize=16)
plt.ylabel('Feature', fontsize=16)
plt.grid()
plt.show()
```

Самые важные признаки - время использования услуг и ежемесечная плата. Их важность в несколько раз выше, чем у остальных признаков.

### Тестирование моделей

Сформируем словарь из моделей

```python
models = {
    'LogisticRegression'     : grid_lr.best_estimator_,
    'RandomForestClassifier' : grid_rf.best_estimator_,
    'LGBMClassifier'         : grid_lgbm.best_estimator_,
    'CatBoostClassifier'     : grid_ctb.best_estimator_
}
```

Обучим на тестовой выборке и вичислим _ROC AUC_ и _accuracy_. Также определим время обучения, точность и полноту

```python
model_results = {}
confusion_matricies = {}

for name, model in tqdm(models.items()):
    res = hm.test_model(model, x_train, x_test, y_train, y_test, accuracy_score)
    y_pred = res.pop('predictions')
    res['accuracy'] = res.pop('score')
    res['AUC-ROC'] = roc_auc_score(y_test, model.predict_proba(x_test)[:,1])
    res['precision'] = precision_score(y_test, y_pred)
    res['recall'] = recall_score(y_test, y_pred)
    model_results[name] = res
    confusion_matricies[name] = confusion_matrix(y_test, y_pred, normalize='true')

hm.roc_curve_plots(models, x_test, y_test)
pd.DataFrame(model_results).T
```

- Лучшая метрика AUC-ROC на тестовой выборке у `LGBMClassifier`. При этом у него также наименьшее время обучения среди всех моделей случайного леса
- У `CatBoostClassifier` AUC-ROC немного ниже, но в то же время самая высокая точность предсказания классов.
- Логистическая регрессия и RandomForestClassifier имеют самые низкие метрики, их не стоит рассматривать

Построим матрицы несоответсвий

```python
fig, ax =plt.subplots(1, 4, figsize=(16,3))
i = 0
for name, matrix in confusion_matricies.items():
    sns.heatmap(pd.DataFrame(matrix), annot=True, cmap='Blues', vmin=0, vmax=1, ax=ax[i])
    ax[i].set(title=name, xlabel='Predictions', ylabel='True')
    i += 1
plt.suptitle('Confusion matricies', fontsize=14, y=1.1)
plt.show()
```

- Логистическая регрессия и CatBoostClassifier хорошо предсказывают отрицательный класс (факт того, что клиент остался), однако хуже всего предсказывает уход клиента и часто дают ложноположительные ответы. В то же время у CatBoostClassifier очень высокая точность
- RandomForestClassifier в среднем ошибается уже меньше. Частота истинных и ложных предсказаний для обоих классов примерно одинакова, но в целом точность очень низкая.
- LGBMClassifier имеет точность на уровне логистической регрессии, чаще остальных моделей правильно предсказывает факт ухода

Предпочтительнее всего использовать LGBMClassifier, так как у него одна из лучших метрик accuracy и лучшая AUC ROC, он быстрее обучается.

### Трудности
Главная трудность возникла при обучении моделей случайного леса, которые оказались чувствительны к способу генерации признака продолжительности использования услуг `UsageTime`:  
Изначально признак `UsageTime` был посчитан как целое число месяцев между датой конца контракта/датой выгрузки и начала контракта. При обучении моделей выяснилось, что модели случайного леса в таком случае были менее склонны замечать разницу между обучающими объектами и выдавали метрики немногим больше, чем логистическая регрессия.

Для решения проблемы этап подготовки и анализа данных был проделан повторно, исследовано влияние разного способа заполнения пропусков, все шаги предшествующие обучению перепроверены. Ошибок не было обнаружено. Проблему решил, заменив способ расчёта `UsageTime` на количество дней. Несмотря на незначительную разницу в форме распределений `UsageTime` в месяцах и в днях использование второго позволяет моделям случайного леса вычленить больше зависимостей.

### Ключевые шаги в решении задачи
- _Добавление признаков_, поскольку датасет был предоставлен без целевой метки. А добавленный признак продолжительности использования услуг оказался наиболее значимым
- _Исследовательский анализ_, поскольку он помогает найти закономерности в поведении клиентов без использовангия машинного обучения и дать некоторые рекомендации заказчику
- _Обучение и анализ моделей_. Анализ склонности поделей давать ложноположительные и ложноотрицательные ответы, позволяет дать рекомендацию по использованию наиболее подходящей модели
